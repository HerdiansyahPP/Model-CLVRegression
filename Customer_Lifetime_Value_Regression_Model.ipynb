# Business Problem Understanding

Auto Insurance adalah sebuah kontrak antara perusahaan asuransi dan pelanggan, dimana pelanggan membayar premi sebagai ganti perlindungan finansial dari kerugian akibat kecelakaan atau pencurian kendaraan, yang kemudian dapat di-klaim oleh pelanggan. Data ini diduga berasal dari perusahaan asuransi di Amerika Serikat, karena kebanyakan negara bagian di AS mewajibkan penduduknya untuk memiliki Basic Auto Insurance. Sebagai pelanggan asuransi, seseorang dapat memilih jenis polis yang sesuai dengan kebutuhan dan anggarannya serta menentukan ketentuan kerusakan yang akan ditanggung oleh perusahaan asuransi. Hal ini akan berpengaruh pada jumlah premi yang harus dibayarkan.

Perusahaan asuransi memperoleh pendapatan dari polis asuransi/premi yang dibayarkan pelanggan, serta dari aktivitas investasi yang dilakukan. Beberapa pelanggan akan membuat klaim dan perusahaan harus membayarkan sesuai dengan polis. Untuk meningkatkan keuntungan, perusahaan asuransi akan berupaya meningkatkan penjualan premi karena tidak dapat mengontrol jumlah klaim. Strategi pemasaran yang tepat dapat meningkatkan penjualan premi dan ada dua fokus dalam pemasaran yaitu mencari pelanggan baru dan mempertahankan pelanggan lama.

Menurut penelitian yang dilakukan oleh [Invesp](https://www.invespcro.com/blog/customer-acquisition-retention/), akuisisi pelanggan memerlukan biaya 5 kali lebih besar dibanding retensi pelanggan. Selain itu, keuntungan yang dapat diperoleh dengan mempertahankan pelanggan lama juga lebih besar karena pelanggan lama cenderung memiliki daya beli yang lebih tinggi dibanding pelanggan baru.


Dikutip dari Glints.com, customer lifetime value atau CLV adalah sebuah metrik atau perhitungan yang penting dalam sebuah bisnis. Metrik ini dapat membantu untuk mengetahui biaya untuk setiap usaha menggaet customer. Secara sederhana, CLV adalah prediksi atas nilai keseluruhan pendapatan yang bisa didapatkan dari pelanggan tersebut.

Berdasarkan [komerce.id](https://komerce.id/blog/customer-lifetime-value/), ada beberapa alasan yang membuat CLV dianggap penting. Berikut ini adalah alasan-alasannya:
- Pertama, pelanggan memiliki nilai penting karena semakin banyak pelanggan, maka semakin besar keuntungan yang akan didapat.
- Kedua, CLV bisa memberi kamu insight berharga tentang seberapa bagus produk kamu dan seberapa baik bisnis terlibat dengan pelanggan.

    **Jika CLV bisnis kamu tinggi, maka itu menunjukkan bahwa bisnis memberi kesan positif pada pelanggan.** 

    Jika rendah, berarti sebaliknya sehingga kamu perlu meluangkan lebih banyak waktu dan usaha untuk meningkatkan produk atau pemasaranmu.
- Ketiga, mengetahui CLV juga dapat meningkatkan laba atas investasi atau juga disebut dengan ROI. Dengan CLV, kamu dapat mengurangi pembelanjaan iklan dengan menargetkan upaya untuk mempertahankan pelanggan yang sudah ada daripada mendapatkan pelanggan baru.

Dilansir dari [glints.com](https://glints.com/id/lowongan/customer-lifetime-value-adalah/#.ZCkb_HZByUk), terdapat beberapa manfaat Customer Lifetime Value:
1. Meningkatkan kesetiaan pelanggan
  
  Secara langsung maupun tidak langsung, kamu akan mengevaluasi pengalaman pelangganmu saat menghitung customer lifetime value. Hasilnya, kamu akan menemukan celah yang masih bisa diperbaiki untuk tingkatkan kepuasan pelanggan.Semakin baik perubahan yang kamu berikan, semakin tinggi juga kesetiaan pelanggan terhadap produk dan brand.

1. Meningkatkan profit perusahaan

  Dilansir dari [Social Head](https://socialhead.io/customer-lifetime-value/#Why_customer_lifetime_value_matters), perusahaan biasanya hanya akan mendapatkan profit sebesar 5-20% dari pelanggan baru dan 60-70% profit dari pelanggan lama. Dari statistik tersebut, dapat kita lihat bahwa CLV adalah salah satu cara untuk mengidentifikasi pelanggan potensial yang berkontribusi besar pada penjualan. Cobalah rumuskan strategi untuk memberikan penawaran terbaik agar mereka mau melakukan repeat purchase. 

1. Memaksimalkan strategi marketing

  Setelah menghitung CLV, kamu akan mengetahui segmentasi konsumen berdasarkan seberapa besar nilai yang mereka berikan pada bisnismu. Segmentasi inilah yang bisa kamu manfaatkan untuk merumuskan strategi marketing yang lebih tepat sasaran. Strategi pemasaran untuk konsumen yang paling sering dan paling jarang membeli produkmu pasti berbeda dan harus dipersonalisasi agar mereka merasa relevan dengan produkmu.

1. Membantu menentukan customer acquisition cost

  Customer acquisition cost adalah biaya yang harus kamu keluarkan untuk menghasilkan pelanggan. Dengan memperhitungkan CLV, kamu akhirnya bisa memprediksi berapa besar biaya tersebut berdasarkan estimasi profit yang akan kamu dapatkan dari mereka nantinya. Hal ini sangatlah penting terutama ketika kamu harus mengalokasikan budget seefisien mungkin.

## Metric Evaluation

Evaluation metrics sangat penting dalam pemodelan karena dapat membantu kita memahami kinerja model dan memudahkan untuk menginterpretasikan model. Kinerja model regresi dapat diketahui dengan mengetahui tingkat error prediksi yang dibuat oleh model tersebut. Model regresi yang baik adalah model yang perbedaan antara nilai aktual dan nilai prediksinya kecil.

Evaluation metrics yang akan digunakan dalam pemodelan ini adalah:

1. RMSE (Root Mean Squared Error): RMSE dihitung dengan meng-akarkuadrat-kan nilai MSE yaitu rata-rata dari kuadrat residual (selisih nilai prediksi dan nilai aktual). Dibandingkan dengan MAE, RMSE akan memberi bobot lebih tinggi pada error sehingga menghasilkan nilai yang lebih besar dibanding MAE. Semakin kecil RMSE akan semakin kecil (mendekati 0) nilai RMSE maka semakin baik. Kita menggunakan metrics ini karena rentang nilai dari Y pada dataset ini tidak terlalu besar (berkisar antara ribuan hingga puluhan ribu).

1. MAE (Mean Absolute Error) : MAE diperoleh dengan cara menghitung nilai rata-rata dari nilai absolute residual (selisih nilai prediksi dan nilai aktual). Semakin kecil nilai MAE, artinya semakin baik pula kinerja model. Kita menggunakan metrics ini karena MAE dapat bekerja dengan baik saat dataset memiliki banyak outlier.

1. R squared : R-Square menjelaskan seberapa besar variasi nilai Y yang dapat dijelaskan oleh model. Nilainya biasanya berkisar antara 0-1, semakin mendekati 1 artinya model semakin baik. Metrics ini digunakan karena lebih mudah dipahami saat menginterpretasikan kinerja model kepada orang lain.

# Data Split

```python
# Define X dan y
X = df_model.drop(columns='Customer Lifetime Value')
y = df_model['Customer Lifetime Value']

# Data Splitting dengan proporsi test size 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state= 0)
```

# Encoding

```python
# Ordinal mapping
ordinal_mapping = [
    {'col':'Coverage', 'mapping':{'Basic':1, 'Extended':2, 'Premium':3, None:0}},
    {'col':'Education', 'mapping':{'High School or Below':1, 'College':2, 'Bachelor':3, 'Master':4, 'Doctor':5, None:0}}
]

# Menggabungkan encoder dalam column transformer
col_transform = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['Vehicle Class', 'EmploymentStatus', 'Renew Offer Type', 'Marital Status']),
    ('ordinal', OrdinalEncoder(cols=['Coverage', 'Education'], mapping=ordinal_mapping), ['Coverage', 'Education'])
], remainder='passthrough')

col_transform
```

Terdapat 6 Kolom Kategorikal yang digunakan:
1. One Hot Encoding: Vehicle Class (6 nunique), EmploymentStatus (5 nunique), Renew Offer Type (4 nunique), Marital Status (3 nunique)
1. Ordinal Encoding: Coverage, Education

# Modeling and Evaluation (10 Feature, 70:30)

```python
# Define scaler
robust = RobustScaler()
```

## Default Model

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state= 0)
rf = RandomForestRegressor(random_state= 0)
ada = AdaBoostRegressor(random_state= 0)
xgb = XGBRegressor(random_state= 0)
gbr = GradientBoostingRegressor(random_state= 0)
lgbm = lgb.LGBMRegressor()

model = [lr, knn, dt, rf, ada, xgb, gbr, lgbm]
```

```python
score_RMSE = []
list_mean_RMSE = []
list_std_RMSE = []
score_MAE = []
list_mean_MAE = []
list_std_MAE = []
score_r2 = []
list_mean_r2 = []
list_std_r2 = []

# Evaluation Metrics yang digunakan
scoring = {'rmse': 'neg_root_mean_squared_error',
           'mae': 'neg_mean_absolute_error',
           'r2': 'r2'}

for i in model:
    pipe_model = Pipeline([
        ('prep', col_transform),
        ('scaler', robust),
        ('algo', i)
    ])

    # Cross validation untuk mencari model terbaik
    cv_score = cross_validate(
        estimator= pipe_model,
        X= X_train, 
        y= y_train,
        scoring= scoring,
        cv = 5
    )

    # RMSE Score
    score_RMSE.append(cv_score['test_rmse'])
    list_mean_RMSE.append(cv_score['test_rmse'].mean())
    list_std_RMSE.append(cv_score['test_rmse'].std())

    # MAE Score
    score_MAE.append(cv_score['test_mae'])
    list_mean_MAE.append(cv_score['test_mae'].mean())
    list_std_MAE.append(cv_score['test_mae'].std())

    # R2 Score
    score_r2.append(cv_score['test_r2'])
    list_mean_r2.append(cv_score['test_r2'].mean())
    list_std_r2.append(cv_score['test_r2'].std())
```

```python
cvresult_before_tuning = pd.DataFrame({
    'model' : ['Linear Regression', 'KNN', 'Decision Tree', 'Random Forest', 'AdaBoost', 'XGBoost', 'Gradient Boost', 'Light GBM'],    
    'mean_RMSE': list_mean_RMSE,
    'std_RMSE': list_std_RMSE,
    'mean_MAE': list_mean_MAE,
    'std_MAE': list_std_MAE,
    'mean_r2': list_mean_r2,
    'std_r2': list_std_r2,
}).round(3)

cvresult_before_tuning
```

```python
cvresult_before_tuning[['model', 'mean_RMSE', 'mean_MAE', 'mean_r2']].sort_values(by=['mean_RMSE','mean_MAE','mean_r2'], ascending=False)
```

Data Validation: GradBoost (RMSE:4010, MAE:1844, r2:0.661); RF (RMSE:4199, MAE:1745, r2:0.627)

```python
# Benchmark 2 model terbaik
test_models = [gbr, rf]

nilai_rmse =[]
nilai_mae =[]
nilai_r2 =[]

# Prediksi pada test set
for i in test_models:
    pipe_model = Pipeline([
        ('prep', col_transform),
        ('scaler', robust),
        ('algo', i)
    ])

    # Fitting & predict
    pipe_model.fit(X_train, y_train)
    yPred = pipe_model.predict(X_test)

    # Menghitung MAE, RMSE, dan MAPE dan digabung dalam list
    nilai_rmse.append(np.sqrt(mean_squared_error(y_test, yPred)))
    nilai_mae.append(mean_absolute_error(y_test, yPred))    
    nilai_r2.append(r2_score(y_test, yPred))

    plt.figure(figsize=(8, 6))
    a = plt.axes(aspect='equal')
    lims = [1000, 80000]
    plt.xlim(lims)
    plt.ylim(lims)
    plt.plot(lims, lims, "r")
    sns.scatterplot(x=y_test, y=yPred).set(title= i, xlabel='Actual CLV', ylabel='Predicted CLV')
```

```python
score_before_tuning = pd.DataFrame({
    'RMSE': nilai_rmse,
    'MAE': nilai_mae,    
    'r2': nilai_r2},
    index=['Gradient Boost', 'Random Forest']).round(3)

score_before_tuning
```

Data Test: RF (RMSE:3940, MAE:1571, r2:0.695); GradBoost (RMSE:3955, MAE:1737, r2:0.693)


Berdasarkan nilai MAE masih cenderung tinggi yaitu sekitar 1571 pada data test. Hal ini dapat diinterpretasikan bahwa model memiliki error sebesar 1571 saat memprediksi nilai CLV. Nilai error ini cenderung besar jika dibandingkan dengan rentang nilai CLV itu sendiri. Dimana nilai CLV bernilai sekitar ribuan sampai puluh ribuan.

Kemudian, hasil visualisasi juga menunjukkan prediksi yang baik untuk nilai CLV di bawah 10.000 namun untuk nilai di atas 10.000 hasil prediksi cenderung tidak tepat dengan nilai sebenarnya (scatterplot cenderung menyebar).

## Default Model

```python
# Define X dan y
X = df_model.drop(columns='Customer Lifetime Value')
y = df_model['Customer Lifetime Value']

# Data Splitting dengan proporsi test size 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state= 0)

# Ordinal mapping
ordinal_mapping = [
    {'col':'Coverage', 'mapping':{'Basic':1, 'Extended':2, 'Premium':3, None:0}},
    {'col':'Education', 'mapping':{'High School or Below':1, 'College':2, 'Bachelor':3, 'Master':4, 'Doctor':5, None:0}}
]

# Menggabungkan encoder dalam column transformer
col_transform = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['Vehicle Class', 'EmploymentStatus', 'Renew Offer Type', 'Marital Status']),
    ('ordinal', OrdinalEncoder(cols=['Coverage', 'Education'], mapping=ordinal_mapping), ['Coverage', 'Education'])
], remainder='passthrough')

# Define scaler
robust = RobustScaler()

# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state= 0)
rf = RandomForestRegressor(random_state= 0)
ada = AdaBoostRegressor(random_state= 0)
xgb = XGBRegressor(random_state= 0)
gbr = GradientBoostingRegressor(random_state= 0)
lgbm = lgb.LGBMRegressor()

model = [lr, knn, dt, rf, ada, xgb, gbr, lgbm]
```

```python
score_RMSE = []
list_mean_RMSE = []
list_std_RMSE = []
score_MAE = []
list_mean_MAE = []
list_std_MAE = []
score_r2 = []
list_mean_r2 = []
list_std_r2 = []

# Evaluation Metrics yang digunakan
scoring = {'rmse': 'neg_root_mean_squared_error',
           'mae': 'neg_mean_absolute_error',
           'r2': 'r2'}

for i in model:
    pipe_model = Pipeline([
        ('prep', col_transform),
        ('scaler', robust),
        ('algo', i)
    ])

    # Cross validation untuk mencari model terbaik
    cv_score = cross_validate(
        estimator= pipe_model,
        X= X_train, 
        y= y_train,
        scoring= scoring,
        cv = 5
    )


    # RMSE Score
    score_RMSE.append(cv_score['test_rmse'])
    list_mean_RMSE.append(cv_score['test_rmse'].mean())
    list_std_RMSE.append(cv_score['test_rmse'].std())

    # MAE Score
    score_MAE.append(cv_score['test_mae'])
    list_mean_MAE.append(cv_score['test_mae'].mean())
    list_std_MAE.append(cv_score['test_mae'].std())

    # R2 Score
    score_r2.append(cv_score['test_r2'])
    list_mean_r2.append(cv_score['test_r2'].mean())
    list_std_r2.append(cv_score['test_r2'].std())
```

```python
cvresult_before_tuning = pd.DataFrame({
    'model' : ['Linear Regression', 'KNN', 'Decision Tree', 'Random Forest', 'AdaBoost', 'XGBoost', 'Gradient Boost', 'Light GBM'],
    'mean_RMSE': list_mean_RMSE,
    'std_RMSE': list_std_RMSE,
    'mean_MAE': list_mean_MAE,
    'std_MAE': list_std_MAE,
    'mean_r2': list_mean_r2,
    'std_r2': list_std_r2,
}).round(3)

cvresult_before_tuning
```

## Hyperparameter Tuning Gradient Boosting

```python
# depth of tree, default=3
max_depth = [3, 4, 7]

# learning rate, default=0.1
learning_rate = [0.5, 0.1, 0.05, 0.01]

# number of tree, default=100
n_estimators = [100, 200, 400]

# Hyperparameter space GBR
hyperparam_space_gbr = {
    'algo__max_depth':max_depth , 
    'algo__learning_rate':learning_rate,
    'algo__n_estimators':n_estimators
}
```

```python
# Membuat algorithm chains
pipe_gbr = Pipeline([
    ('prep', col_transform),
    ('scaler', robust),
    ('algo', gbr)
])

# Hyperparameter tuning dengan GridSearch
grid_gbr = GridSearchCV(
    estimator=pipe_gbr,
    param_grid= hyperparam_space_gbr,
    cv= 5,
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'r2'], 
    refit = 'neg_root_mean_squared_error',
    n_jobs= -1,                          
    verbose= 1,
)
```

```python
grid_gbr.fit(X_train, y_train)
```

Data Train

```python
print('Gradient Boost (by GridSearchCV)')
print('Best_score:', grid_gbr.best_score_)
print('Best_params:', grid_gbr.best_params_)
```

Data Test

```python
# Best_params: {'algo__learning_rate': 0.05, 'algo__max_depth': 3, 'algo__n_estimators': 200}

# Define model terhadap estimator terbaik (via GridSearch)
gbr_tuning = grid_gbr.best_estimator_

# Fitting model
gbr_tuning.fit(X_train, y_train)

# Predict test set
y_pred_gbr_tuning = gbr_tuning.predict(X_test)

# Simpan nilai metrics RMSE, MAE & MAPE setelah tuning
rmse_gbr_tuning = np.sqrt(mean_squared_error(y_test, y_pred_gbr_tuning))
mae_gbr_tuning = mean_absolute_error(y_test, y_pred_gbr_tuning)
r2_gbr_tuning = r2_score(y_test, y_pred_gbr_tuning)

score_after_tuning_gbr = pd.DataFrame({'RMSE': rmse_gbr_tuning, 'MAE': mae_gbr_tuning, 'r2': r2_gbr_tuning}, index= ['Gradient Boost After Tuning']).round(3)
score_after_tuning_gbr
```
